```python
#Imports
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

%config InlineBackend.figure_format ='retina'
```

```python
#Creating Dataset
df = pd.read_csv('../Project/Service_Workplace_Injuries.csv',delimiter=",", low_memory=False)

#Creating Plot
ax = df.plot(x='Amputation', y='Hospitalized', title='Amputated vs Hospitalized', color = ['red'], figsize = (20,20), kind='scatter')
                 
#Add Legend
plt.legend(['Amputation', 'Hospitalized'], loc='upper right', title='Amputated vs Hospitalized')
                 
#Add axes
ax.set_xlabel("Hospitalized")
ax.set_ylabel("Amputation")
                 
#Export the plot
plt.savefig('AmputatedvsHospitalized.pdf')
```
