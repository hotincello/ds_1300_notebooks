```python
#Imports
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

%config InlineBackend.figure_format ='retina'
```

```python
#Creating Dataset
df = pd.read_csv('../Project/Service_Workplace_Injuries.csv',delimiter=",", low_memory=False)

#Creating Plot
ax = df.plot(x='Amputation', y='Hospitalized', title='Amputated vs Hospitalized', color = ['red'], figsize = (20,20), kind='scatter')
                 
#Add Legend
plt.legend(['Amputation', 'Hospitalized'], loc='upper right', title='Amputated vs Hospitalized')
                 
#Add axes
ax.set_xlabel("Hospitalized")
ax.set_ylabel("Amputation")
                 
#Export the plot
plt.savefig('AmputatedvsHospitalized.pdf')
```

```python
import matplotlib.pyplot as plt

# x-coordinates of left sides of bars
left = [1, 2, 3, 4, 5]

# heights of bars
height = [10, 24, 36, 40, 5]

# labels for bars
tick_label = ['Ohio', 'Pennsylvania', 'Florida', 'Texas', 'Illinois']

# plotting a bar chart
plt.bar(left, height, tick_label = tick_label,
		width = 0.8, color = ['red', 'green'])

# naming the x-axis
plt.xlabel('states')
# naming the y-axis
plt.ylabel('hospitalized')
# plot title
plt.title('TOP FIVE States with the most hospitalizations')

# function to show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# defining labels
activities = ['Second degree heat burns', 'Amputations', 'Fractures', 'Lacerations']

# portion covered by each label
slices = [3, 7, 8, 6,]

# color for each label
colors = ['r', 'y', 'g', 'b',]

# plotting the pie chart
plt.pie(slices, labels = activities, colors=colors,
		startangle=90, shadow = True, explode = (0, 0, 0.1, 0),
		radius = 1.2, autopct = '%1.1f%%')

# plotting legend
plt.legend()

# showing the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# x-coordinates of left sides of bars
left = [1, 2, 3, 4, 5]

# heights of bars
height = [10000, 12000, 18000, 25000, 2000]

# labels for bars
tick_label = ['Heat burns', 'Lacerations', 'Amputations', 'Fractures', 'Traumatic injuries']

# plotting a bar chart
plt.bar(left, height, tick_label = tick_label,
		width = 0.2, color = ['blue', 'black'])

# naming the x-axis
plt.xlabel('Injury Type')
# naming the y-axis
plt.ylabel('Number of Cases')
# plot title
plt.title('What are the top five categories of injuries?')

# function to show the plot
plt.show()
```
